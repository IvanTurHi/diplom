```python
import requests
from tqdm import tqdm
```

```python
def get_students_number(session, eoId):
        url = 'https://info-sites.mskobr.ru/api/ekis/classes.json?eoId=' + eoId
        resp = session.get(url)
        #resp = self.session.get(url, headers=self.random_headers())
        js_object = resp.json()
        data = js_object['data']
        primary_number = data['count_pupils_primary']
        secondary_number = data['count_pupils_secondary']
        common_number = data['count_pupils_common']

        return primary_number + secondary_number + common_number
```

```python
session = requests.Session()
students_number_list = []
for i in tqdm(range(schools_df.shape[0])):
  eoid = str(schools_df.iloc[i]['eoid'])
  try:
    snumber = get_students_number(session, eoid)
  except BaseException:
    print('111111111111', eoid)
    snumber = 0
  students_number_list.append(snumber)
students_number_list
```

# Добалвение рейтинга

```python
school_trans_df = gpd.read_file('/content/schools_datamos_final.geojson')
```

```python
site_list = list(set(school_trans_df['website']))
```

```python
for i in site_list:
  sub_df = school_trans_df.loc[school_trans_df['website'] == i].copy(deep=True)
  rating_list = list(sub_df['rating'])
  if len(rating_list) > 0:
    rating = rating_list[0]
    for j in range(len(sub_df)):
      id = sub_df.iloc[j]['id']
      school_trans_df.loc[school_trans_df['id'] == id, 'rating'] = rating
```

```python
school_trans_df.to_file('schools_datamos_final.geojson', driver="GeoJSON")
```

```python
kindergartens_df = gpd.read_file('/content/kindergartens_datamos_final.geojson')
```

```python
site_list = list(set(kindergartens_df['website']))
```

```python
for i in site_list:
  sub_df = kindergartens_df.loc[kindergartens_df['website'] == i].copy(deep=True)
  rating_list = list(sub_df['rating'])

  if len(rating_list) > 0:
    rating_list = [x for x in rating_list if x]
    print(rating_list)
    for j in range(len(rating_list)):
      try:
        rating_list[j] = float(rating_list[j].replace(',', '.'))
      except BaseException:
        rating_list[j] = 4.0
    try:
      avg_rating = round(sum(rating_list) / len(rating_list), 2)
      rating = avg_rating
      for j in range(len(sub_df)):
        id = sub_df.iloc[j]['id']
        kindergartens_df.loc[kindergartens_df['id'] == id, 'rating'] = rating
    except BaseException:
      pass
```

```python
kindergartens_df.to_file('kindergartens_datamos_final.geojson', driver="GeoJSON")
```
