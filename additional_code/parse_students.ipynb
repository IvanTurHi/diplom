```python
number_list = []
list_eoId = []
for i in range(df_school.shape[0]):
  #print(''.join(re.findall(r'\d', df_school.loc[i]['name'])))
  num = ''.join(re.findall(r'\d', df_school.loc[i]['name']))
  list_eoId.append('0')
  if num == '':
    num = '0'
  if len(num) > 4:
    #print(df_school.loc[i])
    num = num[:4]
  number_list.append(num)
  #print(num)
df_school['number'] = number_list
df_school['eoId'] = list_eoId
```
