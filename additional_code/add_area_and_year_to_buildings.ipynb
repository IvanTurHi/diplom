```python
# Преобразование данных с реформы жкх
street_set = set()
street_map = {}
#Извлекаем только улицы из всего адреса
#Лист того что может идти перед названием улицы
prefix_list = ['б-р.', 'пл.', 'кв-л.', 'х.', 'рп.', 'мкр.', 'проезд.', 'аллея.', 'наб.', 'снт.', 'пр-кт.', 'км.', 'туп.', 'ул.', 'пгт.', 'линия.', 'ш.', 'просек.', 'пер.', 'дп.']
for i in range(len(buildings_area_js)):
  for j in prefix_list:
    if j in buildings_area_js[str(i+1)]['address']:
      s = buildings_area_js[str(i+1)]['address'].split(j)[1]
      s = s.split(', ')[0][1:]
      #Костыль тк в некоторых названия указано в скобках название поселка, надо убрать)
      if '(п' in s:
        s = s.split('(п')[0]
        if s[-1] == ' ':
          s = s[:-1]
      street_set.add(s)
      street_map[s] = buildings_area_js[str(i+1)]['address']
  #print(buildings_area_js[str(i+1)]['address'])
street_map
```

```python
def transfrom_address(address):
  space_list = address.split(' ')
  new_s = ''
  new_s += space_list[-1] + ' '
  for j in range(len(space_list)-1):
    new_s += space_list[j] + ' '
  new_s = new_s[:-1]

  return new_s
```

```python
def replace_M_and_B_na_full_name(s, s_for_rep, s_to_rep):
  return s.replace(s_for_rep, s_to_rep)
```

```python
def check_for_M_and_B(address):
  new_s = transfrom_address(address)
  ending = new_s[-2:]
  if 'М.' in new_s:
    s_for_rep = 'М.'
    if ending == 'ий':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Малый')
    elif ending == 'ая':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Малая')
    elif ending == 'ка':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Малая')
    elif ending == 'ый':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Малый')
    elif ending == 'ой':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Малый')
    elif ending == 'ки':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Малые')
    elif ending == 'ин':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Малый')
  if 'Б.' in new_s:
    s_for_rep = 'Б.'
    if ending == 'ий':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Большой')
    elif ending == 'ая':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Большая')
    elif ending == 'ка':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Большая')
    elif ending == 'ый':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Большой')
    elif ending == 'ой':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Большой')
    elif ending == 'ки':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Большие')
    elif ending == 'ин':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Большой')
  return new_s
```

```python
def check_for_Sr_Verh_Nig(address):
  new_s = transfrom_address(address)
  ending = new_s[-2:]
  if 'Ср.' in new_s:
    s_for_rep = 'Ср.'
    if ending == 'ая':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Средняя')
    elif ending == 'ка':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Средняя')
    elif ending == 'ий':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Средний')
    elif ending == 'ый':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Средний')
  if 'Верхн.' in new_s:
    s_for_rep = 'Верхн.'
    if ending == 'ая':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Верхняя')
    elif ending == 'ка':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Верхняя')
    elif ending == 'ий':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Верхний')
    elif ending == 'ый':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Верхний')
  if 'Ниж.' in new_s:
    s_for_rep = 'Ниж.'
    if ending == 'ая':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Нижняя')
    elif ending == 'ка':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Нижняя')
    elif ending == 'ий':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Нижний')
    elif ending == 'ый':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Нижний')
  return new_s
```

```python
def check_for_Star_and_Nov(address):
  new_s = transfrom_address(address)
  ending = new_s[-2:]
  if 'Стар.' in new_s:
    s_for_rep = 'Стар.'
    if ending == 'ий':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Старый')
    if ending == 'ка':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Старая')
    if ending == 'ая':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Старая')
  if 'Нов.' in new_s:
    s_for_rep = 'Нов.'
    if ending == 'ий':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Новый')
    if ending == 'ка':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Новая')
    if ending == 'ая':
      new_s = replace_M_and_B_na_full_name(new_s, s_for_rep, 'Новая')
  return new_s
```

```python
new_street_set = set()
#Ищем в sete конструкцию вида *цифра*-я или *цифра*-й
#Если нашли, и при этом она стоит в конеце(на это второе re тут заточено), то переставляем эту конструкцию в начало
#Если нет конструкции или она и так в начале, то ничего не трогаем
for i in street_set:
  new_s = ''
  list_streets_with_tire = re.findall(r'\d-[я|й]', i)
  if len(list_streets_with_tire) > 0 and ' ' in i:
    if len(re.findall(r'\d-[я|й] ', i)) > 0:
      new_s = i
    else:
      new_s = transfrom_address(i)
  else:
    new_s = i
  #Поиск и замена Б. и М. на Большой и Малый и перенос их из конца адреса в начало
  if len(re.findall(r'[Б|М]\.', i)) > 0:
    new_s = check_for_M_and_B(i)
  #Поиск и замена Ср., Верхн. и Ниж. на Серднюю, Вержнюю и Нижнюю и перенос их из конца адреса в начало
  if len(re.findall(r'Ср\.|Верхн\.|Ниж\.', i)) > 0:
    if 'Нижн.' in i:
      new_s = check_for_Sr_Verh_Nig(i.replace('Нижн.', 'Ниж.'))
    else:
      new_s = check_for_Sr_Verh_Nig(i)
  #Поиск и замена Стар. и Нов. на Старую и Новую и перенос их из конца адреса в начало
  if len(re.findall(r'Стар\.|Нов\.', i)) > 0:
    if 'Ст.' in i:
      new_s = check_for_Star_and_Nov(i.replace('Ст.', 'Стар.'))
    else:
      new_s = check_for_Star_and_Nov(i)
  new_street_set.add(new_s)
  street_map[i] = new_s
new_street_set
```
