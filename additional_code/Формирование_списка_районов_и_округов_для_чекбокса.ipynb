```python
import geopandas as gpd
```

```python
districts_df = gpd.read_file('/content/administrative_borders_transform.geojson')
districts_df
```

#Формируем html для районов

```python
name_to_id_map = {}
for i in range(len(districts_df)):
  name_to_id_map[districts_df.iloc[i]['district_name']] = districts_df.iloc[i]['district_id']
name_to_id_map
```

```python
name_to_id_map_new = {}
for i in name_to_id_map:
  new_key = i.replace('район', '').replace('поселение', '').replace('"', '').strip()
  name_to_id_map_new[new_key] = name_to_id_map[i]
name_to_id_map_new
```

```python
name_to_id_map_new = dict(sorted(name_to_id_map_new.items(), key=lambda x: x[0]))
name_to_id_map_new
```

```python
c = 1
for i in name_to_id_map_new:
  row = """
  <label for="{}"><input type="checkbox" id="{}" name="district{}" value="{}">{}</label>
  """.format(str(c), str(c), str(c), name_to_id_map_new[i], i)
  c += 1
  print(row, end='')
```

#Формируем html для округов

```python
regions_df = gpd.read_file('/content/administrative_regions_transform.geojson')
regions_df
```

```python
name_to_id_map = {}
for i in range(len(regions_df)):
  name_to_id_map[regions_df.iloc[i]['region_name']] = regions_df.iloc[i]['region_id']
name_to_id_map
```

```python
name_to_id_map_new = name_to_id_map
```

```python
name_to_id_map_new = dict(sorted(name_to_id_map_new.items(), key=lambda x: x[0]))
name_to_id_map_new
```

```python
c = 1
for i in name_to_id_map_new:
  row = """
  <label for="r{}"><input type="checkbox" id="r{}" name="region{}" value="{}">{}</label>
  """.format(str(c), str(c), str(c), name_to_id_map_new[i], i)
  c += 1
  print(row, end='')
```
