```python
import geopandas as gpd
import json
```

```python
df_building = gpd.read_file('buildings_raw.geojson')
```

```python
target_list_of_columns = ['id', 'addr:city', 'addr:full', 'addr:housenumber', 'addr:postcode', 'addr:street', 'addr:place', 'building', 'building:levels', 'building:flats', 'design:ref', 'start_date ', 'geometry']
```

```python
#Заполняем пустые значение города Москвой
df_building['addr:city'] = df_building['addr:city'].fillna('Москва')
```

```python
#Заполняем пустые значение посткода нулями
df_building['addr:postcode'] = df_building['addr:postcode'].fillna('000000')
```

```python
#Заполняем поле building на apartments
df_building['building'] = df_building['building'].fillna('apartments')
```

```python
#Заполнение addr:street из addr:place
df_building['addr:street'] = df_building['addr:street'].fillna('empty')
df_building['addr:place'] = df_building['addr:place'].fillna('empty')
for i in range(df_building.shape[0]):
  if df_building.iloc[i]['addr:street'] == 'empty' and df_building.iloc[i]['addr:place'] != 'empty':
    df_building.iloc[i]['addr:street'] = df_building.iloc[i]['addr:place']
```

```python
df_building
```

```python
#Удаление колонок, которые больше не пригодятся
addititonal_dropping_columns = ['addr:full', 'addr:place']
df_building.drop(addititonal_dropping_columns, axis=1, inplace=True)
```

```python
#Заполнение оставшихся пустых данных нулевыми значениями
df_building['building:levels'] = df_building['building:levels'].fillna('0')
df_building['building:flats'] = df_building['building:flats'].fillna('0')
df_building['design:ref'] = df_building['design:ref'].fillna('empty')
```

```python
df_building
```

```python
design_list = list(df_building['design:ref'])
levels_list = list(df_building['building:levels'])
flats_list = list(df_building['building:flats'])
print(len(design_list), len(levels_list), len(flats_list))
print(levels_list.count('0'), flats_list.count('0'))
```

```python
design_to_flat = {}
for i in range(len(design_list)):
  if design_list[i] != 'empty' and design_list[i] not in design_to_flat and flats_list[i] != '0':
    design_to_flat[design_list[i]] = int(flats_list[i])
```

```python
for i in range(len(flats_list)):
  if flats_list[i] == '0' and design_list[i] in design_to_flat:
    flats_list[i] = str(design_to_flat[design_list[i]])
```

```python
design_to_level = {}
for i in range(len(design_list)):
  if design_list[i] != 'empty' and design_list[i] not in design_to_level and levels_list[i] != '0':
    design_to_level[design_list[i]] = int(levels_list[i])
```

```python
design_to_level
```

```python
for i in range(len(flats_list)):
  if levels_list[i] == '0' and design_list[i] in design_to_level:
    levels_list[i] = str(design_to_level[design_list[i]])
```

```python
centroid_list_longitude = []
centroid_list_latitude = []
for i in range(df_building.shape[0]):
  centroid_x = 0
  centroid_y = 0
  counter = 0
  if str(type(df_building.iloc[i]['geometry'])).split("'")[1] == 'shapely.geometry.polygon.Polygon':
    longitude, latitude = df_building.iloc[i]['geometry'].exterior.coords.xy
    for j in range(len(longitude)):
      counter += 1
      centroid_x += longitude[j]
      centroid_y += latitude[j]
    centroid_x = centroid_x / counter
    centroid_y = centroid_y / counter
    
  elif str(type(df_building.iloc[i]['geometry'])).split("'")[1] == 'shapely.geometry.point.Point':
    centroid_x = list(df_building.iloc[-1]['geometry'].coords)[0][0]
    centroid_y = list(df_building.iloc[-1]['geometry'].coords)[0][1]

  elif str(type(df_building.iloc[i]['geometry'])).split("'")[1] == 'shapely.geometry.multipolygon.MultiPolygon':
    mycoordslist = [list(x.exterior.coords) for x in df_building.iloc[i]['geometry'].geoms]
    for j in mycoordslist:
      for k in range(len(j)):
        counter += 1
        centroid_x += j[k][0]
        centroid_y += j[k][1]
    centroid_x = centroid_x / counter
    centroid_y = centroid_y / counter

  elif str(type(df_building.iloc[i]['geometry'])).split("'")[1] == 'shapely.geometry.linestring.LineString':
    longitude, latitude = df_building.iloc[i]['geometry'].coords.xy
    for j in range(len(longitude)):
      counter += 1
      centroid_x += longitude[j]
      centroid_y += latitude[j]
    centroid_x = centroid_x / counter
    centroid_y = centroid_y / counter
    
  centroid_list_longitude.append(centroid_x)
  centroid_list_latitude.append(centroid_y)
```

```python
df_building['centroid longitude'] = centroid_list_longitude
df_building['centroid latitude'] = centroid_list_latitude
```

```python
from pyproj import Geod
from shapely import wkt
import statistics
```

```python
df_building = gpd.read_file('buildings_transform.geojson')
```

```python
border_value = 15000 # Все площади что больше этой заменяются на median_value, подобрано имперически
median_value = 911 # необходимо в двух случаях: когда полигон является точкой и не получается узнать площадь и когда школа выкачивается вместе с окружающей ее территорией, что делает ее площадь во много раз больше. Значение является медианой для площадей школ, площади которых указаны верно
```

```python
area_list = []
geod = Geod(ellps="WGS84")
for i in range(df_building.shape[0]):
  area = 0
  if str(type(df_building.iloc[i]['geometry'])).split("'")[1] == 'shapely.geometry.point.Point':
    area = median_value

  else:
    poly = wkt.loads(str(df_building.iloc[i]['geometry']))
    area = abs(geod.geometry_area_perimeter(poly)[0])
  
  if area > border_value:
    area = median_value
  area_list.append(area)
```

```python
df_building['area'] = area_list
```

```python
#Запись в файл
df_building.to_file('building_transform.geojson', driver="GeoJSON")
```
